```python
from matplotlib import pyplot as pl
import numpy as np
import os
```

```python
def read_energy(name):
    with open(os.path.join(folder1, name), "r") as f:
        d = f.read()
    l = d.split('[')
    data = [[float(j) for j in i[:i.find(']')].split(",")] for i in l[1:]]
    #['mean', 'std', 'min', 'max']
    return data[0]

def read_data(name):
    with open(os.path.join(folder1, name), "r") as f:
        d = f.read()

    l = d.split('[')
    data = [[float(j) for j in i[:i.find(']')].split(",")] for i in l[1:]]
    #['mean', 'max', 'min']
    return data[0], data[1]
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")

folder1 = r"Runs_tourn_size_3_offspring_limited"
files1 = os.listdir(folder1)

energy = [i for i in files1 if 'energy' in i]
data = [i for i in files1 if "data" in i]

mean_energy = []
for name in energy:
    d = read_energy(name)
    mean_energy.append(d)

mean_fitness = []
max_fitness = []
for name in data:
    a, m = read_data(name)
    mean_fitness.append(a)
    max_fitness.append(m)

def amm(val):
    avg = [sum([val[j][i] for j in range(len(val))])/len(val) for i in range(len(val[0]))]
    ma = [np.std([val[j][i] for j in range(len(val))]) + avg[i] for i in range(len(val[0]))]
    mi = [-np.std([val[j][i] for j in range(len(val))]) + avg[i] for i in range(len(val[0]))]
    return avg, ma, mi


dataset = [max_fitness, mean_fitness, mean_energy]
labels = ["Max fitness", "Mean fitness", "Mean energy"]
colors = ['red', "green", "blue"]

fig, ax1 = plt.subplots()

for i in range(3):
    a, ma, mi = amm(dataset[i])
    ax1.plot(a, color=colors[i], label=labels[i])
    ax1.fill_between(range(len(a)), ma, mi, alpha=0.3, color=colors[i])

ax1.legend()
ax1.set_title("EA with tournement size 6", fontsize=18)
ax1.set_xlabel("Generation", fontsize=16)
ax1.set_ylabel("Fitness", fontsize=16)
ax1.set_ylim([-10, 100])

ax2 = ax1.twinx()
ax2.set_ylabel("Energy", fontsize=16)
ax2.set_ylim([-10, 100])

fig.savefig("tourn3.png")
```
